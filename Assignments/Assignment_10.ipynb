```python
"""

1. How do you distinguish between shutil.copy() and shutil.copytree()?

Answer:

shutil.copy - copies single file to destination folder. Destination folder should exist. 
Suppose Destination has the file with the same name, it will be overwritten.

shutil.copytree - copies entire files and directories under source to destination.
(Destination should not already exist)

Example:
           
"""

### shutil.copy

import shutil
import os
print("Folder and Files Under Test1")
print(os.listdir("/Users/mac/Desktop/AI Project/Test1"))
print("\nApplying Shutil.copy")
os.mkdir("/Users/mac/Desktop/AI Project/Test1_Shutilcopy")
des_path= "/Users/mac/Desktop/AI Project/Test1_Shutilcopy"
shutil.copy("new.txt",des_path)
print("\nCopied new.txt from Test1 to Test1_shutilcopy folder")
print("Files under Test1_shutilcopy")
print(os.listdir("/Users/mac/Desktop/AI Project/Test1_Shutilcopy"))

### shutil.copytree 

src_path = "/Users/mac/Desktop/AI Project/Test1"
des_path1 = "/Users/mac/Desktop/AI Project/Test1/Test1_NewBackup"
print("\nFiles and Folders under source")
print(os.listdir(src_path))
print("\nApplying shutil.copytree(Copies all directories and files)")
shutil.copytree(src_path, des_path1)
print("\nFiles and folders under Destination(Copied from source )")
print(os.listdir(des_path1))
```

```python
"""

2. What function is used to rename files??

Answer:

os.rename(old_name,new_name) - This functions is used to rename the files.

Example:

Here i rename multiple files.. see below

"""
from pathlib import Path
path = Path()

old_name = "/Users/mac/Desktop/AI Project/Practice/"
new_name = "/Users/mac/Desktop/AI Project/Practice/"
os.chdir("/Users/mac/Desktop/AI Project/Practice/")
def fileprint():
    for files in path.glob('*.ipynb'):
        print(files)
def filerename():
    count=1
    for files in path.glob('*.ipynb'):
        src_file=old_name+str(files)
        dest_file=new_name+'renamed'+str(count)+'.ipynb'
        os.rename(src_file,dest_file)
        count+=1
def mainfunction():
    print("Files before renaming")
    fileprint()
    print("\n Files after renaming")
    filerename()
    fileprint()
mainfunction() 
```

```python
"""

3. What is the difference between the delete functions in the send2trash and shutil modules?

Answer:

send2trash will move the files or folders to recycle bin.

shutil.rmtree will delete the files or folder permanently.

Example:

"""
import send2trash as stt
import shutil
import os
os.chdir("/Users/mac/Desktop/AI Project/Practice/Test Folder1/")
def fileprint1():
    print("Current Directory is:",os.getcwd())
    print("Files present in this Dir:")
    print(os.listdir())
def mainfunction1():
    print("Before send2trash function")
    fileprint1()
    for files in os.listdir():
        stt.send2trash(files)
    print("\nAfter send2trash function")
    fileprint1()
    print("\nBefore shutil.rmtree()")
    os.chdir("../Test Folder2")
    fileprint1()
    shutil.rmtree(os.getcwd())
    print("\nAfter shutil.rmtree()")
    os.chdir("../")
    fileprint1()
mainfunction1()                
```

```python
"""

4.ZipFile objects have a close() method just like File objects’ close() method. 

What ZipFile method is equivalent to File objects’ open() method?

Answer:

with ZipFile(files,'r') as file - This method is equalant to File Objects open method.

Example below. (Here we open files and extract them all)

"""
import zipfile 
import glob
import os
import pprint
from zipfile import ZipFile
os.chdir("/Users/mac/Downloads/Test Folder1/")
def mainfunction3():
    for files in glob.glob('*.zip'):
        with ZipFile(files,'r') as file:
            file.extractall()
print("Files under Directory")
pprint.pprint(os.listdir(os.getcwd()))
mainfunction3()
print("\n Files after Extracting")
pprint.pprint(os.listdir(os.getcwd()))
```

```python
"""

5. Create a programme that searches a folder tree for files with a certain file extension (such as .pdf or .jpg). 
Copy these files from whatever location they are in to a new folder.

Answer:

"""
import glob
import os
import shutil
import pprint
os.chdir("/Users/mac/Downloads/")
os.mkdir("DbFolder")
src_dest= os.path.join(os.getcwd()+'/DbFolder/')
def mainfunction4():
    print("Db files under Downloads directory")
    for files in glob.glob('*.db'):
        print(files)
        shutil.copy(files,src_dest)
mainfunction4()
print("\nAfter Moving files under new location",src_dest)
pprint.pprint(os.listdir("DbFolder"))
```
