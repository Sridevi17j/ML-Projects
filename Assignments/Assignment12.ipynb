```python
"""
4. If a PdfFileReader object’s PDF is encrypted with the password swordfish, what must you do
before you can obtain Page objects from it?

Answer:

We have to decrypt using decrypt function before doing any sort of work on the pdf.

Example:

"""
import os
import PyPDF2
os.chdir("/Users/mac/Downloads/")
def pdfencryptfun():
    print("opening a file - World Of Learning.pdf")
    file3 = open("World of Learning.pdf",'rb')
    file3_pdfobj = PyPDF2.PdfFileReader(file3)
    numpage=file3_pdfobj.numPages
    file5 = PyPDF2.PdfFileWriter()
    print("Reading the content and storing in new file")
    for i in range(numpage):
        file4 = file3_pdfobj.getPage(i)
        file5.add_page(file4)
    file5.encrypt("PwD")
    with open("World of Learning Encrypted.pdf","wb") as f:
        file5.write(f)
    print("New Encryped File is created with name-World of Learning Encrypted\n")
def pdfdecryptfun():
    file6 = PyPDF2.PdfFileReader("NewPdfFile_encrypted.pdf")
    print("Is PDF File Encrypted?", file6.is_encrypted)
    print("Now we are going to decrypt the file")
    file9 = PyPDF2.PdfWriter()
    if file6.isEncrypted:
        file6.decrypt("PwD")
        for i in range(file6.numPages):
            file8 = file6.getPage(i)
            file9.add_page(file8)
        with open("World of Learning Decrypted.pdf","wb") as f:
            file9.write(f)
        print("Pdf File Decrypted")
    else:
        print("File not decrypted")

pdfencryptfun()
pdfdecryptfun()
```

```python
"""

5. What methods do you use to rotate a page?

Asnwer:

We use rotateClockwise method to rotate page.

Example:

In below example, If we see the pdf files, we can see the actual and rotated one...

"""
    
def loadmodules():
    import os
    import PyPDF2
    os.chdir("/Users/mac/Downloads/")
    
def rotatefun(rotation):
    print("Loaded necessart modules...")
    loadmodules()
    print("opened World of Learning.pdf and reading")
    file10=open("World of Learning.pdf",'rb')
    file11=PyPDF2.PdfFileReader(file10)
    writeobj=PyPDF2.PdfWriter()
    print("\nRotating the file and saving into another file")
    for i in range(file11.numPages):
        page = file11.getPage(i)
        page.rotateClockwise(rotation)
        writeobj.addPage(page)
    file12 = open("World of Learning Rotated.pdf",'wb')
    writeobj.write(file12)
    print("Rotation done, and the new file is - World of Learning Rotated.pdf")
    file10.close()
    file12.close()

rotatefun(180)
```

```python
"""

7. How do you obtain a list of Paragraph objects for a Document object that’s stored in a variable
named doc?

Answer:

we can use doc.paragraphs to obtain list of paragraph objects for a Document object 


"""

import docx

def doc_parafun():
    doc5 = docx.Document()
    doc5_para = doc5.add_paragraph()
    doc5_para.add_run("Hello, this is first paragraph \n").italic= True
    doc5_para.add_run("This is second paragraph in bold").bold = True
    for i in range(len(doc5.paragraphs)):
        print(doc5_para.text)

doc_parafun()
```

```python
"""

8. What type of object has bold, underline, italic, strike, and outline variables?

Answer:

Run object has bold, underline, italic, strike variables

Example:

In below example, if we open the document in word, we can see bold,italic,striked and underlined...

"""
import docx
import os
def runobjfun():
    os.chdir("/Users/mac/Downloads/")
    doc6 = docx.Document()
    doc6_para = doc6.add_paragraph()
    doc6_para.add_run("It is in bold\n").bold = True
    doc6_para.add_run("It is in italic\n").italic=True
    doc6_para.add_run("It is striked\n").font.strike=True
    doc6_para.add_run("It is underlined\n").underline=True
    doc6.save("NewFile.docx")

def fileread(filename):
    filecontent= docx.Document(filename)
    for p in filecontent.paragraphs:
        print(p.text)
    
def main():
    runobjfun()
    fileread("NewFile.docx")
main()
```

```python
"""

11. How do you add a paragraph with the text &#39;Hello, there!&#39; to a Document object stored in a
variable named doc?

Answer:

Using doc.add_paragraph() we can add text, if we want to add style to text, we have to use add_run().

Example:

In below example, if we open the document in word, we can see the styling and color

"""
    
def loadmodules():
    import os
    import docx
    from docx.shared import RGBColor
    os.chdir("/Users/mac/Downloads/")

def docfun():
    import docx
    doc15 = docx.Document()
    doc15_para= doc15.add_paragraph()
    from docx.shared import RGBColor
    run = doc15_para.add_run("&#39;Hello\t")
    run.italic=True
    run.font.color.rgb=RGBColor(124,252,0)
    run = doc15_para.add_run("there!&#39")
    run.italic=True
    run.font.color.rgb=RGBColor(124,252,0)
    doc15.save('TestFile24.docx')
    
def fileread(filename):
    import docx
    filecontent= docx.Document(filename)
    for p in filecontent.paragraphs:
        print(p.text)

def main1():
    loadmodules()
    docfun()
    fileread("TestFile24.docx")

    
main1()
```
