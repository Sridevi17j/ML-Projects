```python
"""

1. Create an assert statement that throws an AssertionError if the variable spam is a negative
integer.

Answer:

"""

print("Program for - Assertion error if variable spam is negative integer")
spam=int(input("Enter Positive Integer"))
try:
    assert spam>0, "Assertion Error:Integer is negative"
    print("your input is",spam)
except AssertionError as e:
    print(e)

```

```python
"""

2. Write an assert statement that triggers an AssertionError if the variables eggs and bacon contain
strings that are the same as each other, even if their cases are different 
(that is, &#39;hello&#39; and &#39;hello&#39; are
considered the same, and &#39;goodbye&#39; and &#39;GOODbye&#39; are also considered the same).

Answer:

"""

def funct1():
    print("Program for Assertion error, if two characters of strings match together")
    eggs = input("Enter Value for Egg Variable")
    bacon = input("Enter Value for bacon variable")
    eggs= eggs.lower()
    bacon=bacon.lower()
    s=[]
    try:
        for i in eggs:
            for j in bacon:
                if i==j:
                    s+=[i]
        s= set(s)
        assert s=={}, "Assertion Error:common characters are :%r "% s
        print("Both strings are different")
    except AssertionError as e:
        print(e)
funct1()                                                                            
```

```python
11. What is the concept of a breakpoint?

Answer:
    
breakpoint() is a built in function. It is used for debugging purpose. 

When we use breakpoint() function it calls, sys.breakpointhook() which calls, pdb.set_trace() function.
So we dont need to explicitly import pdb module.

Also, the environment variable "PYTHONBREAKPOINT" in sys.breakpointhook(), set to 0, then system will execute
code without debugging.

So, having breakpoint() in our program, we can run our program with debugging and without debugging also.

And we can easily change the debugger module. Example if we want to use,
web-pdb debugger, We can easily do it by setting the environment variable PYTHONBREAKPOINT=web_pdb.set_trace.
```
