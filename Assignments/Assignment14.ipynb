```python
"""

5. What method would you call to get Image object for a 100×100 image, excluding the lower-left quarter of it?

Answer:

resize method is used to get an image for a particular size, and here i excluded lower left quarter by turning 

left quarter pixels into black(can be changed to background color also)

Example:

"""

import os
import numpy as np
os.chdir("/Users/mac/Downloads/")
from PIL import Image
Imgobj = Image.open("TestImage.jpg").convert("RGB")
Imgobj=Imgobj.resize((100,100))
img_arr = np.array(Imgobj)
img_arr[50:100 , 0 : 50] = (0, 0, 0)
img_new = Image.fromarray(img_arr)
img_new

```

```python
"""
6. After making changes to an Image object, how could you save it as an image file?

Answer:

.save method will save the image after changes.

Example:

"""

import os
import numpy as np
os.chdir("/Users/mac/Downloads/")
from PIL import Image
Imgobj = Image.open("TestImage.jpg").convert("RGB")
Imgobj=Imgobj.resize((100,100))
img_arr = np.array(Imgobj)
img_arr[50:100 , 0 : 50] = (0, 0, 0)
img_new = Image.fromarray(img_arr)
img_new.save("NewImage.jpeg")

```

```python
"""

7. What module contains Pillow’s shape-drawing code?

Answer:

ImageDraw module of pillow, helps us to draw shapes, add texts in images.

Example:

"""

from PIL import Image, ImageDraw
im = Image.new('RGB', (500, 300), (255, 228, 0))
draw = ImageDraw.Draw(im)
draw.rectangle((100, 100, 150, 200), fill=(255, 0, 0), outline=(0, 0, 0))
im.show()
```

```python
"""

8. Image objects do not have drawing methods. What kind of object does? How do you get this kind of object?

Answer:

Object of ImageDraw.Draw has drawing methods. We can get this object like below.

Example:

"""

from PIL import Image, ImageDraw
im1 = Image.new('RGB', (500, 300), (128, 128, 128))
draw = ImageDraw.Draw(im1)
draw.ellipse((100, 100, 150, 200), fill=(255, 0, 0), outline=(0, 0, 0))
im1.show()
```
