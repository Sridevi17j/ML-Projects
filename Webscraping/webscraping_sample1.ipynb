```python
url = 'https://www.snapdeal.com/products/lifestyle-backpacks?sort=plrty'
```

```python
page = requests.get(url)
```

```python
page
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
soup
```

```python
container = soup.findAll('div',{'class','product-tuple-image'})
```

```python
container
```

```python
productLinks = []
for card in container:
    pLink = card.findAll('a', {'class':'dp-widget-link'})
    pLink = pLink[0]['href']
    productLinks.append(pLink)
```

```python
productLinks
```

```python
with open('snapdealwebscrapping1.csv','w',newline='') as file:
    Writer = csv.writer(file)
    header = ['ProductName','Price']
    Writer.writerow(header)
    for link in productLinks:
        page2 = requests.get(link)
        soup2 = BeautifulSoup(page2.content, 'html.parser')

        productName = soup2.findAll('p', {'class':'product-title'})[0].text
        print(productName)
    
        productPrice = soup2.find('span',{'class':'payBlkBig'}).text
        productPrice = productPrice.replace('₹', '')
        print(productPrice)
        
        #storing each iteration result into csv file
        Writer.writerow([productName, productPrice])
```
